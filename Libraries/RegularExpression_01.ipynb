```python

pattern = r"g..y"
text = "The gray cat is on the tree."

matches = re.findall(pattern, text)
print(matches)  # Output: ['gray']
```

```python

pattern = r"^cat"  # Matches "The" only if it appears at the start of the string.
text = "The cat is on the tree. A cat is cute."

matches = re.findall(pattern, text)
print(matches)
```

```python

pattern = r"tree.$"  # Matches "tall." only if it appears at the end of the string.
text = "The cat is on the tree. A tree is tall."

matches = re.findall(pattern, text)
print(matches)
```

```python

pattern = r"a*c"  # Matches "ac", "abc", "abbc", "abbbc", etc.
text = "ac abc abbc abbbc abbbbc"

matches = re.findall(pattern, text)
print(matches)
```

```python


pattern = r"ab+c"  # Matches "abc", "abbc", "abbbc", etc., but not "ac".
text = "ac abc abbc abbbc abbbbc"

matches = re.findall(pattern, text)
print(matches)
```

```python

pattern = r"ab?c"  # Matches "ac" and "abc", but not "abbc" or others.
text = "ac abc abbc abbbc abbbbc"

matches = re.findall(pattern, text)
print(matches)
```

```python

pattern = r"ab{2,4}c"  # Matches "abbc" and "abbbc", but not "abc" or others.
text = "ac abc abbc abbbc abbbbc"

matches = re.findall(pattern, text)
print(matches)
```

```python

pattern = r"[aeiou]"  # Matches any vowel in the string.
text = "The quick brown fox jumps over the lazy dog."

matches = re.findall(pattern, text)
print(matches)
```

```python

pattern = r"cat|dog"  # Matches either "cat" or "dog".
text = "The cat and the dog are playing in the park."

matches = re.findall(pattern, text)
print(matches)
```

```python
pattern = r"(cat|dog)ing"  # Matches either "cat" or "dog".
text = "The cat and the doge doging are playing in the park."

matches = re.findall(pattern, text)
print(matches)
```

##### \D: Matches any non-digit character (equivalent to [^0-9]).
##### \d: Matches any digit (equivalent to [0-9]).
##### \w: Matches any alphanumeric character and underscore (equivalent to ##### [a-zA-Z0-9_]).
##### \W: Matches any non-alphanumeric character and underscore (equivalent to ##### [^a-zA-Z0-9_]).
##### \s: Matches any whitespace character (space, tab, newline, etc.).
##### \S: Matches any non-whitespace character.
##### \b: Matches the empty string at the beginning or end of a word.
##### \B: Matches the empty string not at the beginning or end of a word.

```python
text = "Age: 25, Height: 180 cm, Weight: 75 kg"

pattern_digits = r"\d+"
digits = re.findall(pattern_digits, text)

print("Digits:", digits)
```

```python
text = "Age: 25, Height: 180 cm, Weight: 75 kg"

pattern_non_digits = r"\D+"
non_digits = re.findall(pattern_non_digits, text)

print("Non-digits:", non_digits)
```

```python
text = "Hello, my name is Eric 5 & and my age is 25."

pattern_non_alphanumeric = r"\W+"
non_alphanumeric = re.findall(pattern_non_alphanumeric, text)

print("Non-alphanumeric:", non_alphanumeric)
```

```python
text = "Hello, \t my name \n # is Jhon."

pattern_whitespace = r"\S+"
whitespace = re.findall(pattern_whitespace, text)

print("Whitespace:", whitespace)
```

```python
text = "Hello, \t my name \n is Usman."

pattern_non_whitespace = r"\S+"
non_whitespace = re.findall(pattern_non_whitespace, text)

print("Non-whitespace:", non_whitespace)
```

```python
text = "The bat is in the battery."

pattern_word_boundary = r"\bbat\b"
word_boundary = re.findall(pattern_word_boundary, text)

print("Word boundary:", word_boundary)
```

```python
text = "The bat is in the battery."

pattern_not_word_boundary = r"\Bbat\B"
not_word_boundary = re.findall(pattern_not_word_boundary, text)

print("Not word boundary:", not_word_boundary)
```

```python
# \d - Matches any digit (equivalent to [0-9])
text_digits = "Age: 25, Height: 180 cm, Weight: 75 kg"
pattern_digits = r"\d+"
digits = re.findall(pattern_digits, text_digits)
print("Digits:", digits)
```

```python
# \D - Matches any non-digit character (equivalent to [^0-9])
text_non_digits = "Age: 25, Height: 180 cm, Weight: 75 kg"
pattern_non_digits = r"\D+"
non_digits = re.findall(pattern_non_digits, text_non_digits)
print("Non-digits:", non_digits)
```

```python
# \w - Matches any alphanumeric character and underscore (equivalent to [a-zA-Z0-9_])
text_alphanumeric = "Hello, my name is Usman and my age is 23."
pattern_alphanumeric = r"\w+"
alphanumeric = re.findall(pattern_alphanumeric, text_alphanumeric)
print("Alphanumeric:", alphanumeric)
```

 In regular expressions, you can use character sets to match any character within the set. Character sets are defined using square brackets [ ]. Here are some examples of

#### [aeiou]: Matches any single lowercase vowel.
#### [A-Za-z]: Matches any single uppercase or lowercase letter.
#### [0-9]: Matches any single digit.
#### [a-zA-Z0-9]: Matches any single alphanumeric character (letter or digit).
#### [^aeiou]: Matches any single character that is not a lowercase vowel.
#### [\s\S]: Matches any single character, including whitespace and non-whitespace  characters.

```python
text = "The quick brown fox jumps over the lazy dog."

pattern_vowel = r"[aeiou]"
vowels = re.findall(pattern_vowel, text)

print("Vowels:", vowels)
```

```python
text = "The quick brown fox jumps over the lazy dog."

pattern_letters = r"[A-Za-z]"
letters = re.findall(pattern_letters, text)

print("Letters:", letters)
```

```python
text = "Age: 25, Height: 180 cm, Weight: 75 kg"

pattern_digit = r"[0-9]"
digits = re.findall(pattern_digit, text)

print("Digits:", digits)
```

```python
text = "Hello, my name is Kamran and my age is 25."

pattern_alphanumeric = r"[a-zA-Z0-9]"
alphanumeric_chars = re.findall(pattern_alphanumeric, text)

print("Alphanumeric characters:", alphanumeric_chars)
```

```python
text = "The quick brown fox jumps over the lazy dog."

pattern_not_vowel = r"[^aeiou]"
not_vowels = re.findall(pattern_not_vowel, text)

print("Characters not vowels:", not_vowels)
```

```python
text = "The quick brown fox jumps over the lazy dog."

pattern_any_character = r"[\s\S]"
any_characters = re.findall(pattern_any_character, text)

print("Any character:", any_characters)
```

```python


def extract_phone_numbers(text):
    # Define the regex pattern for matching phone numbers
    phone_number_pattern = r'\b(?:\d{3}-\d{3}-\d{4}|\(\d{3}\) \d{3}-\d{4})\b'

    # Use re.findall to find all occurrences of the phone number pattern in the text
    phone_numbers = re.findall(phone_number_pattern, text)

    return phone_numbers

# Sample text containing phone numbers
sample_text = """
Here are some phone numbers you can contact:

123-456-7890
(987) 654-3210
555-8675309

Please call any of the above numbers for assistance.
"""

# Call the function to extract phone numbers from the sample text
extracted_phone_numbers = extract_phone_numbers(sample_text)

# Print the extracted phone numbers
print("Extracted phone numbers:")
for phone_number in extracted_phone_numbers:
    print(phone_number)
```
