```python
import numpy as np
```

```python
my_array = np.array([[1,2,3,4,5,6] , [1,2,3,4,5,6]])
print(my_array)
```

```python
my_array[1,5] # to get a specific element from the array you can also give negative value like -1
```

```python
my_array[0 , :] # get a specific rows : means all
```

```python
my_array[: , -2] # to get a specifc column
```

```python
my_array[0 , 1 : 6 : 1]  # statrt , end (this is exclusive) , step size
```

```python
# if you want to change elements in the matrix then you can do it like this
my_array[1,4] = 99
# print(my_array)
#  if you wanna change something like a column then you can specify it like this
my_array[ : ,2] = [0,0]
print(my_array)
```

```python
# 3 dimensional array
b = np.array([[[1,2,3,4,5,6], [1,2,3,4,5,6] , [1,2,3,4,5,6]]])
print(b)
```
