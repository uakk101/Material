```python
import numpy as np
```

```python
# creating array filled with zeros
```

```python
zero_array = np.zeros((4,3))
print(zero_array)
```

```python
# Arrange functions in numpy
arrange_array = np.arange(1,10, 2) # the first argument will be start which is 1n this case the secound argument is where it will stop the array and the last argument will be the steps it will take
print(arrange_array)
```

```python
# Array operation

a = np.array([1,2,3,4,5])
b = np.array([1,2,3,4,5])
result = a + b

print(result)

# this is called concatination of arrays by using plus operator as an operator
```

```python
# Broad casting

a = np.array([[1,2,3,4] , [1,2,3,4]])
b = 10
result = a + b
print(result)
```

```python
sum_array = np.array([1,2,3,4,5,6,7])

print(sum_array.sum())
print(sum_array.mean())
print(sum_array.max())

# NUMPY dot function will do the dot product for two arrays

first_array = np.array([1,2,3,4,5])
secound_array = np.array([1,2,3,4,5])

third_array = np.dot(first_array , secound_array)

print(third_array)
```
